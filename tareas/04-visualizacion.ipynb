```
import altair as alt
import seaborn as sns
```

```
pinguinos = sns.utils.load_dataset("penguins")
```

## __He creado un grafico compuesto que consiste en 3 graficos empalmados.__
## 1er grafico
Este grafico fue de barras. Los cambios fueron que la opacidad se ajustó al 70%, y que se ajustó por color respecto a la especie.
La leyenda de color-especies se eliminó porque resultaba trivial.

## 2do grafico
Este grafico fue de texto, solamente se colocaron las etiquetas de especie dentro de las barras. Para que estas etiquetas se encontraran dentro solamente en vez de usar la función mean(), bastó con usar min(), es decir, en la posición más chica de la longitud de aleta por especie.

## 3er grafico
Este grafico es análogo al 2do, solo que en lugar de colocar la leyenda de especie, se colocó el promedio de las laetas, para que aún mayor la apreciación visual. Es importante recalcar, que se redondeó el promedio a 3 cirfras decimales, esto para que el impacto vsual sea más suave y simétrico.


```
barras = alt.Chart(pinguinos).mark_bar(fillOpacity = 0.70).encode(
    x = alt.X('mean(flipper_length_mm):Q', title = 'Longitud promedio de aletas'),
    y = alt.Y('species:N', title = 'Especie'),
    color = alt.Color('species:N', legend = None)
)


etiquetas_especies = alt.Chart(pinguinos).mark_text(size = 15).encode(
    x = alt.X('min(flipper_length_mm):Q'),
    y = alt.Y('species:N'),
    text = alt.Text('species')
)

etiquetas_mean = alt.Chart(pinguinos).mark_text(size=13, dy=13).encode(
    x = alt.X('min(flipper_length_mm):Q'),
    y = alt.Y('species:N'),
    text = alt.Text('mean(flipper_length_mm):Q', format = '.3f'),
)

grafico_compuesto = alt.layer(barras, etiquetas_especies, etiquetas_mean).properties(title = 'Aletas en diferentes especies de pinguinos').properties(
    height = 250,
    width = 600,
)

grafico_compuesto
```

```
grafico_compuesto.save('mi_grafico_con_altair.html')
```
